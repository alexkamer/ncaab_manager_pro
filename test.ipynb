```python
import sqlite3

def view_database_schema(db_file):
    """Connects to an SQLite database and prints the schema of all tables."""
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        cursor = conn.cursor()

        # Query the sqlite_master (or sqlite_schema) table to get all table names
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
        tables = cursor.fetchall()

        if not tables:
            print(f"No tables found in the database: {db_file}")
            return

        print(f"--- Schema for database: {db_file} ---")

        # Iterate over each table and print its schema
        for table_name_tuple in tables:
            table_name = table_name_tuple[0]
            print(f"\n*** Table: {table_name} ***")
            # Query sqlite_master again to get the CREATE TABLE SQL statement
            cursor.execute("SELECT sql FROM sqlite_master WHERE name=?", (table_name,))
            create_sql = cursor.fetchone()[0]
            print(create_sql)

    except sqlite3.Error as e:
        print(f"An error occurred: {e}")
    finally:
        if conn:
            conn.close()

# Specify your database file name
database_file = 'data/ncaab.db'
view_database_schema(database_file)
```

```python
import pandas as pd
import sqlite3

def load_db_table_to_dataframe(db_file, table_name):
    """
    Connects to an SQLite database and loads a specified table into a pandas DataFrame.
    """
    conn = None
    try:
        # Establish a connection to the SQLite database
        conn = sqlite3.connect(db_file)
        print(f"Connection established to {db_file}")

        # Construct the SQL query to select all data from the table
        sql_query = f"SELECT * FROM {table_name}"

        # Use pandas to read the SQL query results directly into a DataFrame
        df = pd.read_sql_query(sql_query, conn)
        
        print(f"Successfully loaded '{table_name}' into a DataFrame.")
        print(f"DataFrame shape: {df.shape}")
        
        return df

    except sqlite3.Error as e:
        print(f"An error occurred: {e}")
        return None
    finally:
        # Ensure the connection is closed
        if conn:
            conn.close()
            print("Database connection closed.")

# --- Example Usage ---

# Assume you have a table named 'teams' in your ncaab.db
database_file = 'data/ncaab.db'
table_to_load = 'player_boxscores' # Replace 'teams' with the actual name of your table

# Assign the table data to a pandas variable
teams_df = load_db_table_to_dataframe(database_file, table_to_load)

# You can now work with the pandas DataFrame
if teams_df is not None:
    print("\nFirst 5 rows of the DataFrame:")
    print(teams_df.head())

    print("\nDataFrame columns and data types:")
    print(teams_df.info())
```
